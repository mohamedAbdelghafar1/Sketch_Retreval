```python
import os
import shutil
import random


base_dir = 'C:/Users/20127/Desktop/sketch_retrieval/dataset_mine/dataset_mine/new_updated_categories/'

# Get list of class directories
class_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

# Define paths

train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')




for class_dir in class_dirs:
    # Create class subdirectories in train and validation directories
    os.makedirs(os.path.join(train_dir, class_dir), exist_ok=True)
    os.makedirs(os.path.join(validation_dir, class_dir), exist_ok=True)
    
    # Get list of images in the current class directory
    class_path = os.path.join(base_dir, class_dir)
    images = os.listdir(class_path)
    
    # Shuffle images and split into training and validation sets
    random.shuffle(images)
    split_idx = int(len(images) * 0.9)
    train_images = images[:split_idx]
    validation_images = images[split_idx:]
    
    # Move images to train and validation directories
    for image in train_images:
        shutil.move(os.path.join(class_path, image), os.path.join(train_dir, class_dir, image))
    
    for image in validation_images:
        shutil.move(os.path.join(class_path, image), os.path.join(validation_dir, class_dir, image))
    
    # Remove the original class directory if empty
    if not os.listdir(class_path):
        os.rmdir(class_path)

print('Dataset reorganized successfully.')
```
